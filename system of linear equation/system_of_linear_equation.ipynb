## **Problem 1**

Make a program for visualizing the solution of a 3x3 system of linear equations problem

$A \bar{x} = \bar{b}, \; \; 
\bar{x} \in \mathbb{R}^{3}, \;
\bar{b} \in \mathbb{R}^{3}$

When
1. $\operatorname{rank} (A) = 3$
2. $\operatorname{rank} (A) = 2$
3. $\operatorname{rank} (A) = 1$

### **Code**

````
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline  

def visualize_systems(A, b, view1, view2):
    # Check rank of matrix
    rank = np.linalg.matrix_rank(A)
    print(f"rank(A)       = {rank}")

    # find solution if rank = 3
    if (rank==3):
        x_sol = np.dot(np.linalg.inv(A), b)
        print(f"solution of x = {x_sol}")

    # Create the meshgrid
    x, y = np.linspace(-4, 4, 100), np.linspace(-4, 4, 100)
    X, Y = np.meshgrid(x,y)

    # Calculate
    Z1 = (b[0] - A[0][0]*X - A[0][1]*Y)/A[0][2]
    Z2 = (b[1] - A[1][0]*X - A[1][1]*Y)/A[1][2]
    Z3 = (b[2] - A[2][0]*X - A[2][1]*Y)/A[2][2]
    
    # Create figure
    fig = plt.figure(figsize=(22, 8))
    stride = 25

    # Figure 1
    ax = fig.add_subplot(121, projection='3d')
    ax.view_init(view1[1], view1[0])

    ax.plot_surface(X, Y, Z1, facecolor='r', alpha=.2, rstride=stride, cstride=stride)
    ax.plot_surface(X, Y, Z2, facecolor='g', alpha=.2, rstride=stride, cstride=stride)
    ax.plot_surface(X, Y, Z3, facecolor='b', alpha=.2, rstride=stride, cstride=stride)

    ax.plot_wireframe(X, Y, Z1, rstride=stride, cstride=stride)
    ax.plot_wireframe(X, Y, Z2, rstride=stride, cstride=stride)
    ax.plot_wireframe(X, Y, Z3, rstride=stride, cstride=stride)
    if (rank == 3):
        ax.scatter((x_sol[0],),(x_sol[1],),(x_sol[2],), c='red', s=150)
    ax.set_title("View angle - 1")

    plt.xlabel('x')
    plt.ylabel('y')

    # Figure 2
    ax2 = fig.add_subplot(122, projection='3d')
    ax2.view_init(view2[1], view2[0])

    ax2.plot_surface(X, Y, Z1, facecolor='r', alpha=.2, rstride=stride, cstride=stride)
    ax2.plot_surface(X, Y, Z2, facecolor='g', alpha=.2, rstride=stride, cstride=stride)
    ax2.plot_surface(X, Y, Z3, facecolor='b', alpha=.2, rstride=stride, cstride=stride)

    ax2.plot_wireframe(X, Y, Z1, rstride=stride, cstride=stride)
    ax2.plot_wireframe(X, Y, Z2, rstride=stride, cstride=stride)
    ax2.plot_wireframe(X, Y, Z3, rstride=stride, cstride=stride)
    if (rank == 3):
        ax2.scatter((x_sol[0],),(x_sol[1],),(x_sol[2],), c='red', s=150)
    ax2.set_title("View angle - 2")

    plt.xlabel('x')
    plt.ylabel('y')

    plt.show()
````

### **Problem 2-1**

In **Gauss-Jordan Elimination** procedure, we create an augmented matrix and solve the lower triangular matrix.

<img src="https://i.ytimg.com/vi/UhhfKXBnnEo/maxresdefault.jpg" width="50%"/>

Below is the **Gauss-Jordan elemination** procedure

<img src="https://www.codewithc.com/wp-content/uploads/2014/05/gauss-elimination-flowchart-pivoting-and-elimination.png" />

#### **Code**

````
import numpy as np
import sys

def gauss_jordan_elimination(A, b):
    # 1. Arrange augmented matrix
    augmented_matrix = np.concatenate((A, b[:, None]), axis=1).astype(np.float64)

    # 2. start eliminating
    (n_columns, n_rows) = augmented_matrix.shape   
    x = np.zeros(n_columns).astype(np.float64)
    for i in range(n_columns-1):
        # check whether a_ii = 0
        if (augmented_matrix[i][i] == 0.0):
            sys.exit("Error - diagonal element zero!")

        # if not, eliminate the row
        for j in range(i+1, n_columns):
            ratio = float(augmented_matrix[j][i] / augmented_matrix[i][i])
            augmented_matrix[j] = augmented_matrix[j] - ratio*augmented_matrix[i]
        
        print(f"Augmented Matrix - iteration-{i+1}")
        print(f"{augmented_matrix}")
        print("")

    
    # 3. Obtain solution
    for i in range(n_columns):
        sum = 0
        for j in range(i):
            sum += x[n_columns-j-1]*augmented_matrix[n_columns-i-1][n_columns-j-1]
        x[n_columns-i-1] = (augmented_matrix[n_columns-i-1][3] - sum)/ \
                            (augmented_matrix[n_columns-i-1][n_columns-i-1])

    return (x)
````

### **Problem 2-2**

To compute the inverse of matrix $A$, we create the augmented matrix by concatenating matrix of identity.
<br><br>
$\begin{bmatrix}
2 & 1 & -1 & | & 1 & 0 & 0\\ 
-3 & -1 & 2 & | & 0 & 1 & 0\\ 
-2 & 1 & 2 & | & 0 & 0 & 1
\end{bmatrix}$
<br><br>
Then, we perform the previous **Gauss-Jordan Elimination** and continue the procedure by calculating the upper triangle matrix. After that, we normalize the matrix to obtain the inverse.
<br><br>
The latest **Gauss-Jordan Elimination** step
<br><br>
$\begin{bmatrix}
1 & 0 & 0 & | & a & b & c\\ 
0 & 1 & 0 & | & d & e & f\\ 
0 & 0 & 1 & | & g & h & i
\end{bmatrix}$
<br><br>
$\text{inv}(A) = \begin{bmatrix}
a & b & c\\ 
d & e & f\\ 
g & h & i
\end{bmatrix}$

#### **Code**

````
def gauss_jordan_elimination_inverse(A):
    # 1. Arrange augmented matrix
    augmented_matrix = np.concatenate((A, np.identity(len(A))), axis=1).astype(np.float64)
    #print("Initial augmented matrix")
    #print(augmented_matrix)
    #print("")

    # 2. start eliminating
    (n_columns, n_rows) = augmented_matrix.shape   
    
    # create lower triangle
    for i in range(n_columns-1):
        # check whether a_ii = 0
        if (augmented_matrix[i][i] == 0.0):
            sys.exit("Error - diagonal element zero!")

        # if not, eliminate the row
        for j in range(i+1, n_columns):
            ratio = float(augmented_matrix[j][i] / augmented_matrix[i][i])
            augmented_matrix[j] = augmented_matrix[j] - ratio*augmented_matrix[i]
        
        #print(f"Lower triangle - iteration-{i+1}")
        #print(augmented_matrix)
        #print("")

    # create upper triangle
    for i in range(n_columns):
        for j in range(i):
            ratio = augmented_matrix[n_columns-i-1][n_columns-j-1]/ \
                    augmented_matrix[n_columns-j-1][n_columns-j-1]

            augmented_matrix[n_columns-i-1] = augmented_matrix[n_columns-i-1] - \
                                                ratio*augmented_matrix[n_columns-j-1]
            
            

        augmented_matrix[n_columns-i-1] = augmented_matrix[n_columns-i-1]/ \
                        augmented_matrix[n_columns-i-1][n_columns-i-1]
        #print(f"Upper triangle - iteration-{i+1}")
        #print(f"{augmented_matrix}")
        #print("")

    _, inverse_matrix = np.hsplit(augmented_matrix, 2)
    
    return (inverse_matrix)
````

### **Problem 2-3**

To compute the determinant of $A$, we can use the **Gauss-Jordan Elimination** procedure from **Problem 2-1**.
<br><br>
The latest **Gauss-Jordan Elimination** step is:

$\begin{bmatrix}
a & b & c  \\ 
0 & d & e  \\ 
0 & 0 & f 
\end{bmatrix}$
<br><br>
Then, the determintant of $A$ is

$\text{det}(A) = a.d.f$

#### **Code**

````
def gauss_jordan_elimination_determinant(A):
    # 1. Arrange augmented matrix
    augmented_matrix = A.astype(np.float64)

    # 2. start eliminating
    (n_columns, n_rows) = augmented_matrix.shape   
    for i in range(n_columns-1):
        # check whether a_ii = 0
        if (augmented_matrix[i][i] == 0.0):
            sys.exit("Error - diagonal element zero!")

        # if not, eliminate the row
        for j in range(i+1, n_columns):
            ratio = float(augmented_matrix[j][i] / augmented_matrix[i][i])
            augmented_matrix[j] = augmented_matrix[j] - ratio*augmented_matrix[i]
    
    # 3. Obtain solution
    det = 1
    for i in range(n_columns):
        det *= augmented_matrix[i][i]

    return (det)
````

### **Problem 2-4**

Solve the problem using **Cramer's Rule**

<img src="https://media.cheggcdn.com/media%2F569%2F56987a75-0863-462c-905b-ef4d323226bb%2FphpWzUpO4.png" 
width="50%" />


#### **Code**

````
def cramer_rules(A, b):
    # Set the matrix
    list_of_D = []

    for i in range(len(A)+1):
        temp_Mat = A.copy()
        if (i<len(A)):
            for j in range(len(A)):
                temp_Mat[j][i] = b[j]
        
        list_of_D.append(np.linalg.det(temp_Mat))

    # Obtain the solution
    x = np.zeros(len(A))
    for i in range(len(x)):
        x[i] = list_of_D[i]/list_of_D[-1]

    return (x)
````
